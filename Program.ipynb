```python
df = pd.read_excel("Dataset_Apriori_6_Kolom.xlsx.")
data = df.drop(columns=["ID_Transaksi"])
jumlah_transaksi = len(data)
print("Jumlah Transaksi:", jumlah_transaksi)
data.head()
```

```python
min_support = 0.2
support_1 = {}

for col in data.columns:
    support = data[col].sum() / jumlah_transaksi
    if support >= min_support:
        support_1[col] = support

support_1
```

```python
support_2 = {}

for itemset in itertools.combinations(support_1.keys(), 2):
    count = 0
    for _, row in data.iterrows():
        if row[itemset[0]] == 1 and row[itemset[1]] == 1:
            count += 1

    support = count / jumlah_transaksi
    if support >= min_support:
        support_2[itemset] = support

support_2
```

```python
min_confidence = 0.6
rules = []

for (A, B), supp_ab in support_2.items():
    conf_A_B = supp_ab / support_1[A]
    conf_B_A = supp_ab / support_1[B]

    if conf_A_B >= min_confidence:
        rules.append((A, B, conf_A_B))

    if conf_B_A >= min_confidence:
        rules.append((B, A, conf_B_A))

rules
```
